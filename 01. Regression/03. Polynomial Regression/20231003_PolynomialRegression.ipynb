### 03.- Regresión polinomial

### a) Importación de las librerías

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### b) Importación del conjunto de datos

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

### c) Entrenamiento del modelo de regresión lineal en todo el conjunto de datos

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

### d) Entrenamiento del modelo de regresión polinomial en todo el conjunto de datos

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

### d) Visualización de los resultados de la regresión lineal

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Regresión Lineal')
plt.xlabel('Nivel de la posición')
plt.ylabel('Salario')
plt.show()
```

### e) Visualización de los resultados de la regresión polinomial

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Regresión Polinomial')
plt.xlabel('Nivel de la posición')
plt.ylabel('Salario')
plt.show()
```

### f) Visualización de los resultados de la regresión polinomial (para una resolución más alta y una curva más suave)

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Regresión Polinomial')
plt.xlabel('Nivel de la posición')
plt.ylabel('Salario')
plt.show()
```

### g) Predecir un nuevo resultado con regresión lineal

```python
lin_reg.predict([[6.5]])
```

### h) Predecir un nuevo resultado con regresión polinomial

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
