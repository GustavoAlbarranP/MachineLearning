```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
```

```python
tf.__version__
```

### Parte 1 - Preprocesamiento de datos

### a) Preprocesamiento del conjunto de entrenamiento

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
training_set = train_datagen.flow_from_directory('dataset/training_set',
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode = 'binary')
```

### b) Preprocesamiento del banco de pruebas

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
test_set = test_datagen.flow_from_directory('dataset/test_set',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode = 'binary')
```

## Parte 2 - Construir la CNN

### a) Inicialización de la CNN

```python
cnn = tf.keras.models.Sequential()
```

### Paso 1 - Convolución

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[64, 64, 3]))
```

### Paso 2 - Puesta en común

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

### b) Añadir una segunda capa convolucional

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

### Paso 3 - Aplanado

```python
cnn.add(tf.keras.layers.Flatten())
```

### Paso 4 - Conexión completa

```python
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
```

### Paso 5 - Capa de salida

```python
cnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

## Parte 3 - Entrenamiento de la CNN

### a) Compilación de la CNN

```python
cnn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

### b) Entrenamiento de la CNN en el conjunto de entrenamiento y evaluación en el conjunto de prueba

```python
cnn.fit(x = training_set, validation_data = test_set, epochs = 25)
```

## Parte 4 - Hacer una única predicción

```python
import numpy as np
from keras.preprocessing import image
test_image = image.load_img('dataset/single_prediction/cat_or_dog_1.jpg', target_size = (64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)
result = cnn.predict(test_image)
training_set.class_indices
if result[0][0] == 1:
  prediction = 'dog'
else:
  prediction = 'cat'
```
