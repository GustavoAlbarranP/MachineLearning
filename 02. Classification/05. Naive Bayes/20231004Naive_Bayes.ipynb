## 05.- Naive Bayes

### a) Importación de las librerías

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### b) Importación del conjunto de datos

```python
dataset = pd.read_csv('SNA.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

### c) Dividir el conjunto de datos en el conjunto de entrenamiento y el conjunto de prueba

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

### d) Escalado de características

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

### e) Entrenamiento del modelo Naive Bayes en el conjunto de entrenamiento

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

### f) Predecir un nuevo resultado

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

### g) Predicción de los resultados del conjunto de pruebas

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### h) Crear la matriz de confusión

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

### i) Visualizar los resultados del conjunto de entremaniemto

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)
plt.title('Naive Bayes (Training set)')
plt.xlabel('Edad')
plt.ylabel('Salario Estimado')
plt.legend()
plt.show()
```

### j) Visualizar los resultados del conjunto de test

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)
plt.title('Naive Bayes (Training set)')
plt.xlabel('Edad')
plt.ylabel('Salario Estimado')
plt.legend()
plt.show()
```

El algoritmo de Naive Bayes es un clasificador probabilístico basado en el teorema de Bayes, con la "ingenua" suposición de independencia entre las características. Es particularmente adecuado cuando la dimensión de los datos es alta. A pesar de su simplicidad, Naive Bayes puede superar a clasificadores más sofisticados. Aquí está la explicación paso a paso:

### 1. Teorema de Bayes
La base del algoritmo es el teorema de Bayes, que es una forma de calcular la probabilidad posterior \( P(Y|X) \) a partir de \( P(Y) \), \( P(X) \) y \( P(X|Y) \):
\[ P(Y|X) = \frac{P(X|Y) \cdot P(Y)}{P(X)} \]

### 2. Suposición de Independencia
Naive Bayes simplifica el cálculo asumiendo que las características son condicionalmente independientes dado el resultado. Por lo tanto, la probabilidad conjunta se puede expresar como el producto de las probabilidades individuales:
\[ P(X|Y) = P(x_1|Y) \cdot P(x_2|Y) \cdot ... \cdot P(x_n|Y) \]

### 3. Modelo de Probabilidad
Para un conjunto de datos de entrenamiento, se calcula la probabilidad de cada clase \( P(Y) \) y la probabilidad condicional de cada característica dado cada clase \( P(x_i|Y) \).

### 4. Calculando Probabilidades
- **Probabilidad de Clase:** \( P(Y) \) se calcula como la frecuencia relativa de la clase \( Y \) en el conjunto de entrenamiento.
- **Probabilidad Condicional:** \( P(x_i|Y) \) se calcula como la frecuencia de la característica \( x_i \) en muestras de la clase \( Y \).

### 5. Suavizado de Laplace
Para evitar la multiplicación por cero cuando una característica no ha sido observada con una clase en los datos de entrenamiento, se utiliza el suavizado de Laplace añadiendo 1 a los conteos de las frecuencias.

### 6. Clasificación
Para clasificar una nueva instancia, se calcula la probabilidad posterior para cada clase y se selecciona la clase con la mayor probabilidad posterior.

### 7. Trabajar con Datos Continuos
En el caso de características continuas, se asume una distribución (normalmente una distribución gaussiana) y se calculan los parámetros de esa distribución (media y desviación estándar) para cada clase.

### 8. Decisiones en la Vida Real
Aunque se calcula \( P(X) \) para el teorema de Bayes, en la práctica este término se omite ya que es constante para todas las clases y no afecta la decisión final.

### Ejemplo Práctico:
Supongamos que tenemos un conjunto de datos de correos electrónicos y queremos clasificarlos en "spam" o "no spam". Calculamos las probabilidades de cada palabra dada la clase (spam/no spam) y la probabilidad de cada clase. Luego, para un nuevo correo electrónico, multiplicamos las probabilidades de cada palabra observada (asumiendo independencia) y clasificamos el correo electrónico en la categoría con mayor probabilidad posterior.

Naive Bayes es un algoritmo poderoso por su simplicidad, eficiencia y efectividad, especialmente en tareas de clasificación de texto como filtrado de spam y análisis de sentimientos.
