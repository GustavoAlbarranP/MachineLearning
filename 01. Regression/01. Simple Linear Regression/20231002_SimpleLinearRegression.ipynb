## 01.- Regresión Lineal Simple

### a) Importar las librerías

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### b) Importar los datasets

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

### c) Dividir el dataset entre conjuntos de entrenamiento y testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

### d) Entrenar el modelo de Regresión Lineal Simple usando el conjunto de entrenamiento

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

### e) Predecir los resultados de test

```python
y_pred = regressor.predict(X_test)
```

### f) Visualizar los resultados en el conjunto de entrenamiento

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Sueldo vs Experiencia (Conjunto de Entrenamiento)')
plt.xlabel('Años de Experiencia')
plt.ylabel('Sueldo')
plt.show()
```

### g) Visualizar los resultados en el conjunto de testing

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Sueldo vs Experiencia (Conjunto de Test)')
plt.xlabel('Años de Experiencia')
plt.ylabel('Sueldo')
plt.show()
```

La regresión lineal simple es un método estadístico que permite estudiar y modelar la relación entre dos variables continuas: una variable dependiente $Y$ y una variable independiente $X$. Aquí está la descripción paso a paso:

### 1. Modelo Lineal
En regresión lineal simple, se modela la relación entre las dos variables con una línea recta:
$Y = \beta_0 + \beta_1X + \varepsilon$
donde:
- $\beta_0$ es la ordenada al origen (el valor de $Y$ cuando $X = 0$).
- $\beta_1$ es la pendiente de la línea (el cambio en $Y$ por cada unidad de cambio en $X$).
- $\varepsilon$ es el término de error que captura cualquier variación en $Y$ no explicada por $X$.



### 2. Estimación de Parámetros
Los parámetros $\beta_0$ y $\beta_1$ se estiman a partir de los datos. La estimación más común es el método de mínimos cuadrados, que busca minimizar la suma de los cuadrados de los errores (las diferencias entre los valores observados y los valores predichos por el modelo).

### 3. Mínimos Cuadrados
La solución de mínimos cuadrados viene dada por las siguientes fórmulas:
$$\beta_1 = \frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n}(x_i - \bar{x})^2}$$
$$\beta_0 = \bar{y} - \beta_1\bar{x}$$
donde $\bar{x}$ y $\bar{y}$ son las medias de las variables independiente y dependiente respectivamente.

### 4. Ajuste del Modelo
Una vez que tenemos los estimadores $\hat{\beta_0}$ y $\hat{\beta_1}$, podemos ajustar la línea de regresión a los datos y usarla para hacer predicciones.

### 5. Evaluación del Modelo
Evaluamos qué tan bien el modelo se ajusta a los datos usando varias métricas, como el coeficiente de determinación $R^2$, que mide la proporción de la variabilidad en $Y$ que puede ser explicada por $X$ a través del modelo.

### 6. Supuestos
El modelo de regresión lineal simple se basa en varios supuestos, incluyendo:
- Linealidad: la relación entre $X$ y $Y$ es lineal.
- Independencia: los errores $\varepsilon$ son independientes.
- Homoscedasticidad: la varianza de los errores es constante para todos los niveles de $X$.
- Normalidad: los errores $\varepsilon$ siguen una distribución normal.

### 7. Diagnóstico del Modelo
Para confirmar que el modelo es adecuado, se realizan pruebas de diagnóstico, como análisis de residuos, para verificar los supuestos y detectar posibles problemas como la heteroscedasticidad o la no linealidad.

Supongamos que queremos estudiar la relación entre la temperatura (variable independiente $X$) y el consumo de helados (variable dependiente $Y$). Recolectamos datos sobre la temperatura y las ventas de helados, estimamos $\beta_0$ y $\beta_1$ usando mínimos cuadrados, y usamos nuestro modelo para predecir las ventas de helados a diferentes temperaturas.

La regresión lineal simple es fundamental en estadística y es a menudo el punto de partida para el análisis de relaciones entre variables.
